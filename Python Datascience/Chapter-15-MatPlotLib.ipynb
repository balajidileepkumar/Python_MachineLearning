```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)

t2 = np.arange(0.0, 5.0, 0.05)
#print(t1,"\n", f(t1),"\n",t2,"\n",f(t2))

#plt.figure(1)
plt.subplot(211)#Row/Column/position
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
 
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'k',t2, np.cos(2*np.pi*t2), 'gD')
plt.show()
     
```

```python
def sintheta(t):
    return np.sin(t)
def costheta(t):
    return np.cos(t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(5.0, 0.0, -0.05)
#print(t1,"\n", f(t1),"\n",t2,"\n",f(t2))

#plt.figure(1)
plt.subplot(211)
plt.plot(t1, sintheta(t1), 'bo', t1, costheta(t1), 'k')
 
plt.subplot(212)
plt.plot(t2, sintheta(t2), 'k',t2, costheta(t2), 'gD')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Multiple Subplots on mathematical functions with subplot adjustments
from matplotlib.ticker import NullFormatter  # useful for `logit` scale
    
# Fixing random state for reproducibility
np.random.seed(100)
    
# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
#Generates a normal distribution of numpy array with loc as a mean and scale as standard deviation 
#print(y)

#loc : float or array_like of floats
#Mean (centre) of the distribution.

#scale : float or array_like of floats
#Standard deviation (spread or width) of the distribution.

#size : int or tuple of ints, optional
#Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn. If size is None (default), a single value is returned if loc and scale are both scalars. Otherwise, np.broadcast(loc, scale).size samples are drawn.

#print(y)
print(len(y))
#print(y>0)
#print([(y > 0) & (y < 1)])
y = y[(y > 0) & (y < 1)]
#print(len(y))
#print(y)
y.sort(axis=0)
#print(y[::-1])
x = np.arange(len(y))
#print(x)   
# plot with various axes scales
fig = plt.figure(1)
fig, ax = plt.subplots() 
 
# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
  
  
# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
  
  
#symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.05)
plt.title('symlog')
plt.grid(True)
  
# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=2, bottom=0, left=0.0, right=2, hspace=0.8,
               wspace=0.5)
size = fig.get_size_inches() # size in pixels
sizeindpi = size*fig.dpi
print(size, sizeindpi)

print("Dot per inch(DPI) for the figure is: ", fig.dpi)
bbox = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
width, height = bbox.width, bbox.height
print("Axis sizes are(in pixels):", width, height)
plt.show()
#print(plt.gca())
#print(plt.gca().yaxis)
#print(plt.gca().xaxis)


```

```python
import matplotlib
from scipy.stats import gamma
distr = gamma(2.0)
xvalues = np.linspace(0, 10, 100)
yvalues = distr.pdf(xvalues)
plt.plot(xvalues, yvalues, color='orange')
plt.title('Mean and median')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
xmean = distr.mean()
ymean = distr.pdf(xmean)
xmed = distr.median()
ymed = distr.pdf(xmed)
print(xmean,ymean,xmed,ymed)
#aprops = dict(arrowstyle = '|-|',connectionstyle=matplotlib.patches.ConnectionStyle.Arc3(rad=0.5))#<|-|>
aprops = dict(arrowstyle = '->',connectionstyle=matplotlib.patches.ConnectionStyle.Angle3(angleA=90, angleB=0))#<|-|>
#aprops = dict(arrowstyle = '|-|',connectionstyle=matplotlib.patches.ConnectionStyle.Bar(armA=0.0, armB=0.0, fraction=0.3, angle=None))#<|-|>
#aprops = dict(arrowstyle = 'fancy')#<|-|>

plt.annotate('Median={:3.2f}'.format(xmed), 
             (xmed, ymed), 
             (xmed + 2, ymed+ 0.05),
             arrowprops=aprops)
ymean = distr.pdf(xmean)

#print(xmean,ymean,xmed,ymed)
plt.annotate('Mean={:3.2f}'.format(xmean), 
             (xmean, ymean), 
             (xmean+1, ymean+0.03),
             arrowprops=aprops)
plt.grid()
comment = '''In a skewed distribution, 
mean and median usually
have different values.'''
plt.text(4.5, 0.15, comment, fontsize=8, color='blue')
None
```
