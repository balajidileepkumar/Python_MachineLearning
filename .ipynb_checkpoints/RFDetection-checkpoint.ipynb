```python
import pandas as pd
import scipy as sp
import os
```

```python
# Set some pandas options
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)
```

```python
PathRF="RF.csv"
PathDMI="DMI.csv"
PathSOI="SOI.csv"
if(os.path.exists(PathRF)):
    global DF_RF,DF_DMI,DF_SOI
    DF_RF = pd.read_csv(PathRF, index_col=['Year'])
    DF_RF.drop(DF_RF.columns[DF_RF.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
    print(DF_RF.head(10))
    print("############################################")
if(os.path.exists(PathDMI)):
    DF_DMI = pd.read_csv(PathDMI, index_col=['Year'])
    DF_DMI.drop(DF_DMI.columns[DF_DMI.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
    print(DF_DMI.head(10))
    print("############################################")
if(os.path.exists(PathSOI)):
    DF_SOI = pd.read_csv(PathSOI, index_col=['Year'])
    DF_SOI.drop(DF_SOI.columns[DF_SOI.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
    print(DF_SOI.head(10))
    print("############################################")
```

```python
#DF_RF_MONTHS = DF_RF[0,1]
DF_RF_MONTHS = pd.MultiIndex.from_frame(DF_RF)

print(DF_RF_MONTHS)
RF_MI_SERIES = DF_RF.stack()
print(type(RF_MI_SERIES))
print(RF_MI_SERIES)
```

```python
#RF_MI_RAINS = DF_RF.filter(items=['Total','Rainy Days'],usecols=["YEARLY_RAINFALL","TOTAL_RAINY_DAYS"])
RF_MI_RAINS = DF_RF.filter(items=['Total','Rainy Days'])
RF_MI_RAINS_SERIES = RF_MI_RAINS.stack()
print(RF_MI_RAINS)
#RF_MI_RAINS_SERIES
```

```python
DMI_MI_SERIES = DF_DMI.stack()
print(type(DMI_MI_SERIES))
print(DMI_MI_SERIES)
```

```python
SOI_MI_SERIES = DF_SOI.stack()
print(type(SOI_MI_SERIES))
print(SOI_MI_SERIES)
```

```python
DF_list = [DF_RF,DF_DMI,DF_SOI]
DF_Combined = pd.concat(DF_list)
print(DF_Combined)
#DF_Combined.columns
```

```python
DF_COMBINED = pd.DataFrame({'RAINFALL_MM': RF_MI_SERIES, 'DIPOLE_MODE_INDEX': DMI_MI_SERIES, 'SOUTHERN_OSCILLATION_INDEX':SOI_MI_SERIES})
print(DF_COMBINED)
```
