```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame , Series
df=pd.read_csv("P1-OfficeSupplies.csv")
#print(df)
df["Total"]=df.Units * df["Unit Price"]
#print(df)
x=df[(df.Region == "East")]
#print (x)
y=df[(df.Region == "Central")]
#print(y)
z=df[(df.Region == "West")]
#print(z
a=(x.Total.sum())
b=(y.Total.sum())
c=(z.Total.sum())
labels='EAST','CENTRAL','WEST'
sizes=[a,b,c]

plt.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title("Total sales in Region",fontsize=40)
plt.show()
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
sales=df.Total
rep=df.Rep
ax.bar(rep,sales,color='g')
plt.xlabel("Rep",fontsize=20)
plt.ylabel("Total sales",fontsize=20)
plt.title("Total Sales by Rep",fontsize=40)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse
import math

df=pd.read_csv("P1-OfficeSupplies.csv")
unit_price = df["Unit Price"]
no_units = df["Units"]
rep_name = df["Rep"]

#ells = [Ellipse(xy=rnd.rand(2)*10, width=rnd.rand(), height=rnd.rand(), angle=rnd.rand()*360) for i in range(NUM)]
ells = [Ellipse(xy=rnd.rand(2)*10, width=math.sqrt(x[2]), height=math.sqrt(x[1]), angle=rnd.rand()*360) for x in zip(rep_name,unit_price,no_units)]

fig = plt.figure(20)
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

plt.show()
```
