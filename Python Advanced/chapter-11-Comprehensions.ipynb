## Comprehensions
### List Comprehensions & Dictionary Comprehensions

In addition to sequence operations and list methods, Python includes a more advanced operation called a list comprehension.

List comprehensions allow us to build out lists using a different notation. You can think of it as essentially a one line <code>for</code> loop built inside of brackets. For a simple example:


## Example 1

```python
mystring = 'helloworld'
```

```python
mylist = []
mylist_1 = list()

for letter in mystring:
    if(letter not in 'aeiou'):
        mylist.append(letter)
print(mylist)
```

```python
for charposition in range(0,len(mystring),2):
    if(charposition +1 < len(mystring)):
        mylist_1.append(mystring[charposition]+mystring[charposition +1])
    else:
        mylist_1.append(mystring[charposition])
print(mylist_1)
```

### Dictionary Comprehension

```python
a = {"fruit1": "apple", "fruit2":"orange", "fruit3":"pomegrante", "fruit4":"grape" }
ab = {print(x) if(len(x) > 6) else print("replace ", x) for x in a.values()}
print(ab)
```

```python
a = {"fruit1": "apple", "fruit2":"orange", "fruit3":"pomegrante", "fruit4":"grape" }
{print(x) if(len(x) > 6) else print("replace ", x) for x in a.values()}
a = {20: "apple", 15:"orange", 8:"pomegrante", 7:"grape" }
print("############################")
{print(x[0]) if(x[0] > 10) else print("shortage ", 10 - x[0]) for x in a.items()}
print("############################")
result ={}
{result.setdefault(x[1], 0) if(x[0] > 10) else result.setdefault(x[1], 10 - x[0]) for x in a.items()}
print(result)
```
